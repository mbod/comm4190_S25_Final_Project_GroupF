# University City Dining Navigator: RAG-Based Recommendation System

## Project Concept and Innovation

Our project develops a context-aware restaurant recommendation system for the University City area using Retrieval-Augmented Generation (RAG) technology. Unlike conventional recommendation engines that rely solely on ratings or keywords, our system understands the nuanced situations diners face and delivers targeted suggestions that account for real-world constraints and social contexts.

The innovation lies in our approach to dining recommendations as situational rather than purely preference-based. By understanding the "why" behind a restaurant search—whether it's a quick meal between classes, an impressive date venue, or a group gathering—we deliver more meaningful and useful results than traditional platforms like Yelp or Google Reviews.

## Key User Scenarios

Our research identified three primary dining decision scenarios that University City community members regularly face:

### 1. The Time-Constrained Diner

This scenario addresses users who need immediate food options but want to avoid repeating recent dining choices. Key considerations include:
- Proximity to current location  
- Estimated service/wait time  
- Variety from recently visited establishments  
- Quick-service capabilities  
- Grab-and-go options  

### 2. The Impression-Focused Experience

This scenario serves users planning dining experiences where the venue itself carries social significance:
- Relationship context (romantic date, professional meeting, family visit)  
- Atmosphere requirements (intimate, sophisticated, relaxed)  
- Budget parameters and perceived value  
- Cuisine appropriateness for the occasion  
- Reservation capabilities and availability  

### 3. The Group Consensus Challenge

This scenario helps resolve the complex decision-making process when multiple people with different preferences need to dine together:
- Accommodation of diverse dietary needs  
- Space and seating configurations for groups  
- Menu variety to satisfy multiple preferences  
- Noise levels appropriate for group conversation  
- Cost considerations across varying budgets  

## Technical Implementation

### Data Architecture

Our implementation diverges from existing approaches by creating a custom-structured dataset specifically designed for situational dining recommendations:

**1. Data Collection Method**
- Manual extraction and verification of restaurant information from multiple sources  
- Structured categorization beyond standard metrics (noise levels during different times, typical clientele, hidden features not mentioned in main listings)  
- Regular updates to maintain accuracy  

**2. Data Attributes (Beyond Standard Information)**
- Contextual tags (good for groups, impressive for dates, quick service)  
- Temporal factors (busy periods, typical wait times, peak noise hours)  
- Spatial considerations (walking distance from campus landmarks, accessibility)  
- Social factors (typical customer demographics, attire expectations)  

**3. Metadata Enhancement**
- Sentiment analysis from review text to extract subjective qualities  
- Identification of "hidden gems" versus "popular standards"  
- Extraction of time-sensitive information (seasonal menus, special events)  

### RAG System Architecture

Our RAG implementation connects our specialized restaurant dataset with natural language understanding to interpret situational dining needs:

**1. Query Understanding Layer**
- Context classification to identify which user scenario is being addressed  
- Constraint extraction to identify non-negotiable factors  
- Preference weighting to balance competing priorities  

**2. Retrieval Mechanism**
- Situation-based vector embedding to connect scenarios with appropriate venues  
- Dynamic re-ranking based on contextual importance  
- Personalization through user interaction history  

**3. Response Generation**
- Explanation-rich recommendations that provide rationale for each suggestion  
- Handling of follow-up questions and refinement requests  
- Alternative suggestion generation when initial recommendations aren't satisfactory
